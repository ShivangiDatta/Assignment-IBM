```python
import pandas as pd
df = pd.read_csv('heart_2020_cleaned.csv')
```

```python
df.head()
```

```python
df['HeartDisease'] = df['HeartDisease'].map({'Yes': 1.0, 'No': 0})
df
```

```python
df.columns
```

##Data Preparation



```python
y = df['HeartDisease']
y
```

```python
X = df.drop(['HeartDisease', 'PhysicalHealth', 'MentalHealth','DiffWalking', 'Race', 'PhysicalActivity', 'GenHealth', 'Asthma', 'SkinCancer', 'AgeCategory', 'Sex', 'Diabetic'], axis=1)
X
```

```python
X.head(5)
```

```python
X['Smoking'] = X['Smoking'].map({'Yes': 1.0, 'No': 0})
X['AlcoholDrinking'] = X['AlcoholDrinking'].map({'Yes': 1.0, 'No': 0})
X['Stroke'] = X['Stroke'].map({'Yes': 1.0, 'No': 0})
X['KidneyDisease'] = X['KidneyDisease'].map({'Yes': 1.0, 'No': 0})
X
```

##Data Splitting

```python

import numpy as np
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
X_train
```

# **Model Building**

##Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```
